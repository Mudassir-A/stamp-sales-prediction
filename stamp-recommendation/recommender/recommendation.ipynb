```python
import pandas as pd
import numpy as np
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import nltk

# Download required NLTK data
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')

# Initialize lemmatizer and stop words
lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))

def preprocess_text(text):
    """
    Preprocess text by:
    1. Converting to lowercase
    2. Tokenizing
    3. Removing stop words and punctuation
    4. Lemmatizing
    """
    # Convert to lowercase
    text = str(text).lower()
    
    # Tokenize
    tokens = word_tokenize(text)
    
    # Remove stop words and punctuation, then lemmatize
    tokens = [lemmatizer.lemmatize(token) for token in tokens 
             if token.isalnum() and token not in stop_words]
    
    return tokens

# Read the CSV file
df = pd.read_csv('stamps_data_21to24.csv')

# Create new column with tokenized words
df['tokenized_name'] = df['name'].apply(preprocess_text)

# Remove specified columns
columns_to_remove = ["sl_no", "printer", "fdc_images", "brochure_pdf"]
df = df.drop(columns=columns_to_remove)

# Display first few rows
print("\nFirst few rows of the processed dataframe:")
print(df.head())

# Display shape of the dataframe
print("\nShape of the dataframe:", df.shape)

# Display sample of tokenized names
print("\nSample of tokenized names:")
for name, tokens in zip(data['name'].head(), df['tokenized_name'].head()):
    print(f"\nOriginal: {name}")
    print(f"Tokens: {tokens}")
```

```python
def get_stamp_recommendations(df):
    """
    Get stamp recommendations based on user input
    """
    # Display available stamps for reference
    print("\nAvailable stamps:")
    for idx, name in enumerate(df['name'], 1):
        print(f"{idx}. {name}")
    
    # Get user input
    stamp_name = input("\nEnter the name of the stamp you want recommendations for: ")
    
    # Tokenize the input stamp name using the same preprocessing
    input_tokens = set(preprocess_text(stamp_name))
    
    # Calculate relevance score for each stamp
    def calculate_relevance(row_tokens):
        row_tokens = set(row_tokens)
        common_words = len(input_tokens.intersection(row_tokens))
        return common_words / len(input_tokens) if len(input_tokens) > 0 else 0
    
    # Add relevance score column
    df['relevance_score'] = df['tokenized_name'].apply(calculate_relevance)
    
    # Filter and sort recommendations
    recommendations = df[df['relevance_score'] > 0].sort_values(
        by='relevance_score', ascending=False
    )
    
    # Remove the exact match if present
    recommendations = recommendations[
        recommendations['name'].str.lower() != stamp_name.lower()
    ]
    
    return recommendations[['name', 'release_date', 'denomination', 'quantity', 'relevance_score']]

# Interactive recommendation system
while True:
    # Get and display recommendations
    recommendations = get_stamp_recommendations(df)
    print("\nTop recommendations:")
    print(recommendations.head())
    
    # Ask if user wants to continue
    continue_search = input("\nWould you like to search for another stamp? (yes/no): ").lower()
    if continue_search != 'yes':
        break

print("\nThank you for using the stamp recommendation system!")
```
